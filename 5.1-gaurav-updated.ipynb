```python
#distribution of total coupon accepted per coupon type
sns.countplot(data = data, x = "coupon", hue = 'Y', order=data['coupon'].value_counts(ascending=True).index)
plt.title('Count of Coupons accepted- by coupon type', size = 10)
plt.xticks(rotation=45)
```

```python
#NOTE- max coupon accepted are of type Carry out & Take away and Restaurant (< 20)
```

```python
#total and percentage of accepted coupons

#create dataframe for Some college - no degree
for coupon_type in ['Restaurant(<20)','Coffee House','Carry out & Take away','Bar','Restaurant(20-50)']:
   grp1 = data[data['coupon'].isin([coupon_type])]
   val= ((grp1['Y'].value_counts()[1])/len(grp1.index))*100
   print("Percentage of coupons accepted with coupon type-",coupon_type,val) 
    
```

```python
#NOTE- coupon with category Carry out & Take away and Restaurant(<20) are accepted with 70%+
```

```python
#create new dataframe with bar data only
dfbar = (data[data['coupon']=='Bar'])
#need to find the number of redeemed coupons and divide by the total number of coupons issued for this population.
percent = ((dfbar['Y'].value_counts()[1])/len(dfbar.index))
print("Percentage of total bar coupon accepted : {:,.1%}".format(percent))
```

```python
#create dataframe for 3 or fewer bar visits
dfbar_3few = dfbar[dfbar['Bar'].isin(['never','less1','1~3'])]
#calculate the percent of redeemed
percent_redeemed = dfbar_3few['Y'].value_counts(normalize=True).get(1, 0) * 100

#calculate the percent of not redeemed
percent_not_redeemed = 100 - percent_redeemed
print("Total observations of Bar visitors (3 or Fewer visits/month) accepting coupons: {:.1f}%".format(percent_redeemed))
print("Total observations of Bar visitors (3 or Fewer visits/month) not accepting coupons: {:.1f}%".format(percent_not_redeemed))

#calculate var between not accepting vs accepting
variance_percent = (percent_not_redeemed - percent_redeemed) / percent_redeemed
print("Variance between not accepting and accepting the coupon by Bar visitors (3 or Fewer visits/month): {:.1%}".format(variance_percent))
```

```python
bar_more1 = ["1~3", "4~8", "gt8"]
gend = ['Male']

# create column for above conditions
#data[((data['Bar'].isin(bar_more1)) & (~data['age'].isin(age_less25)), True, False)]
df2 = data[data['Bar'].isin(bar_more1) & data['gender'].isin(gend)]
male_fewer = df2['Y'].value_counts()[1]/len(df2.index) * 100
print("Total observations of Male Bar visitors (3 or Fewer visits/month) accepting coupons is {:.1f}%".format(male_fewer))
```

```python
bar_more1 = ["1~3", "4~8", "gt8"]
age_less25 = ['21', 'below21']

# create column for above conditions
#data[((data['Bar'].isin(bar_more1)) & (~data['age'].isin(age_less25)), True, False)]
df = data[data['Bar'].isin(bar_more1) & ~data['age'].isin(age_less25)]
age25_bar = df['Y'].value_counts()[1]/len(df.index)*100
print("Total observations  vistor with age more than 25 and Bar visitors (3 or Fewer visits/month) accepting coupons is {:.1f}%".format(age25_bar))
```

```python
#NOTE- Drivers over 25 years and go to the bar more than once a month are more likely to accept the coupon 62% times
```
